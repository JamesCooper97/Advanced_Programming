```python
# JSON to XML
# constructing the tree node by node
# this code is specific to the data's structure
# it is not a genric solution for all XML docs
import json
import xml.etree.ElementTree as ET

inFile = open("People.json")
dataFile = json.load(inFile)
#print(dataFile)

newRoot = ET.Element('students')

for student in dataFile['students']:
    studentElm = ET.SubElement(newRoot, 'student', {'id':str(student['id'])})
    fullElm = ET.SubElement(studentElm, 'fullName',)
    titleElm = ET.SubElement(fullElm, 'title')
    titleElm.text = student['fullName']['title']
    firstElm = ET.SubElement(fullElm, 'first')
    firstElm.text = student['fullName']['first']
    surnameElm = ET.SubElement(fullElm, 'surname')
    surnameElm.text = student['fullName']['surname']
    
    if student['fullName']['other'] != 'null':
        for other in student['fullName']['other']:
            otherElm = ET.SubElement(fullElm, 'other')
            otherElm.text = other
            
    ageElm = ET.SubElement(studentElm, 'age')
    ageElm.text = str(student['age'])
    cityElm = ET.SubElement(studentElm, 'city')
    cityElm.text = str(student['city'])
#ET.dump(newRoot)
newTree = ET.ElementTree(newRoot)
newTree.write("people2.xml")
```

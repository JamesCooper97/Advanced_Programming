## Exercise one
Using an appropriate Python data structure, implement two programs that demonstrate the behaviour of a stack (LIFO) queue (FIFO). For each program, you should create a menu system to enable the user to manipulate the structure, and the program should keep running until the user chooses to exit. An example menu for a stack would be:
```
1. Push
2. Pop
3. View
```

### Definitions:
LIFO - Last In First Out\
FIFO - First In First Out

### Stack (LIFO) commands:
Push - Adds an element to the top of the stack
Pop - Removes the topmost element of the stack, aka the most recently added element
[[1]](#1).

### Queue (FIFO) commands:
Push - Adds an element to the back of the queue
Pop - Removes the element at the front of the queue
[[2]](#2)

<a id="1">[1]</a> Safa Mulani & Bradley Kouchi, How to Implement a Stack in C Programming, DigitalOcean, www.digitalocean.com. https://www.digitalocean.com/community/tutorials/stack-in-c (accessed Nov. 1, 2024)
<a id="2">[2]</a> Queue push() and pop() in C++ STL, GeeksForGeeks.org, https://www.geeksforgeeks.org/queue-push-and-queue-pop-in-cpp-stl (accessed Nov. 1, 2024)

````python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)
        print(f"{item} is appended to stack")

    def pop(self):
        if not self.stack:
            print("Stack empty")
        else:
            item = self.stack[0]
            self.stack.pop()
            print(f"{item} popped from stack")

    def view(self):
        if not self.stack:
            print("Stack empty")
        else:
            print(self.stack)

def main():
    stack = Stack() # initialise a stack
    print("Stack Menu:")
    print("1 - push")
    print("2 - pop")
    print("3 - view")
    print("4 - exit")
    while True:

        choice = input ("Enter an integer between and including 1 and 4, then hit enter")

        if choice == '1':
            item = input("Enter the item to push: ")
            stack.push(item)
        elif choice == '2':
            stack.pop()
        elif choice == '3':
            stack.view()
        elif choice == '4':
            print("Exiting the stack program.")
            break
        else:
            print("Invalid choice. Please try again.")
        
        

if __name__ == "__main__":
    main()
````

````python
class Queue:
    def __init__(self):
        self.queue = []

    def push(self, item):
        self.queue.append(item)
        print(f"{item} is appended to queue")

    def pop(self):
        if not self.queue:
            print("Queue empty")
        else:
            item = self.queue[0]
            self.queue.pop(0)
            print(f"{item} popped from queue")

    def view(self):
        if not self.queue:
            print("Queue empty")
        else:
            print(self.queue)

def main():
    queue = Queue() # initialise a queue
    print("Queue Menu:")
    print("1 - push")
    print("2 - pop")
    print("3 - view")
    print("4 - exit")
    while True:
        choice = input ("Enter an integer between and including 1 and 4, then hit enter")

        if choice == '1':
            item = input("Enter the item to push: ")
            queue.push(item)
        elif choice == '2':
            queue.pop()
        elif choice == '3':
            queue.view()
        elif choice == '4':
            print("Exiting the queue program.")
            break
        else:
            print("Invalid choice. Please try again.")
        
        

if __name__ == "__main__":
    main()
````

## Exercise two
A business has a number of retail outlets across the world. Each outlet keeps a track of its monthly sales and sends a quarterly report to the company's headquarters. Using at least two different data structures, write a program that calculates the average for each outlet and each month, the total for each outlet and each month, and the grand total for the quarter. You should hard code (write data into your program directly) all the data you need once only. The output should be formatted as shown below.

Hint: Remember, a data structure can be nested within another data structure, and the content of one data structure can be used as a reference point in another (see the sample code).

Sales by month table

| ---- | April 18 | May 18 | June 18 |
|------| -------- | ------ | ------- |
| L3   | 390      | 345    | 379     |
| P2   | 250      | 270    | 300     |
| N6   | 460      | 480    | 450     |
| B8   | 470      |  510   | 360     |

Sales for the second quarter in $1000's

L3=London, P2=Paris, N6=New York, B8=Beijing

Quarterly sales report

```
London        L3:   $Average   $Total
Paris         P2:   $Average   $Total
New York      N6:   $Average   $Total
Beijing       B8:   $Average   $Total

Apr18           :   $Average   $Total
May18           :   $Average   $Total
Jun18           :   $Average   $Total

Total: $GrandTotal
```

````python
sales = {
    "L3": {"name": "London", "sales": [390, 345, 379]},
    "P2": {"name": "Paris", "sales": [250, 270, 300]},
    "N6": {"name": "New York", "sales": [460, 480, 450]},
    "B8": {"name": "Beijing", "sales": [470, 510, 360]}
}

month_list = ["Apr18","May18","Jun18"]

month_totals = [0] * len(month_list)

def main():
    print("Quarterly Sales Report\n")

    grand_total = 0
    
    for outlet_key, outlet in sales.items():
        outlet_total_sales = sum(outlet["sales"])
        
        outlet_sales_average = outlet_total_sales/len(outlet["sales"])
        grand_total += outlet_total_sales

        print(f"{outlet['name']:12} {outlet_key}: ${outlet_sales_average:.2f} ${outlet_total_sales}")

        for i, outlet_month_sale in enumerate(outlet["sales"]):
            month_totals[i] += outlet_month_sale

    print("")

    for i, month in enumerate(month_list):
        month_average = month_totals[i] / len(sales)
        print(f"{month:15}: ${month_average:.2f} ${month_totals[i]}")

    print(f"\nTotal: ${grand_total}")
        

if __name__ == "__main__":
    main()

        
````
